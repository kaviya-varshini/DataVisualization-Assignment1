```python
import matplotlib.pyplot as plt
import numpy as np

# Use numpy to generate a bunch of random data in a bell curve around 5.
n = 5 + np.random.randn(1000)

m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Raw Data")
plt.show()

plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

```python
data.plot(kind="scatter",x="Age",y="Fare")
```

```python
sns.boxplot(x="Age",y="Fare",data=data)
```

```python
sns.violinplot(x="Age",y="Fare",data=data,size=6)
plt.show()
```

```python
sns.FacetGrid(data,hue="SibSp",size=5).map(sns.kdeplot,"Pclass").add_legend()
plt.show()
```

```python
sns.barplot(x="Age",y="Fare",data=data,hue="SibSp")
plt.show()
```

```python
labels = ["Age","Fare","SibSp"]
sizes = [50,50,50]
plt.pie(sizes,labels=labels,explode=(0.1,0.1,0.1))
plt.axis("equal")
plt.show()
```
